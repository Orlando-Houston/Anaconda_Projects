```python
# see the variables with the highest 5 correlations between them
for col in boston.feature_names:
    highest_values=abs(boston_X.corr()[col]).nlargest(n=5)
    print(highest_values)
    #print if value greater than 0.75.
    for index,value in highest_values.items():
        if 1>value >= 0.75:
            print(index,col,'variables are highly correlated: ',value)
```
