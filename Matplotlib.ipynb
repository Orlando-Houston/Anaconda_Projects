```python
plt.plot(x,np.sin(x),label='sinus');
plt.plot(x,np.cos(x),label ='cosinus');
plt.legend()
```

```python
plt.plot(x,np.sin(x),label='sinus',linestyle='--');
plt.plot(x,np.cos(x),label ='cosinus',linestyle ='-.');
plt.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,color = "red",marker ="o",markersize=20
        ,markerfacecolor ="black",markeredgecolor = "yellow",markeredgewidth = 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,lw=3,ls ="--")
ax.set_xlim(0,20)
ax.set_ylim(0,50)
```
