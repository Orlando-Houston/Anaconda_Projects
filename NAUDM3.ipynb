```python
import numpy as np
import pandas as pd
```

## Create random dataframe

```python
df = pd.DataFrame({'a':np.random.randint(10,10000,10),'b':np.random.randint(10,10000,10)})
df
```

## Simple manhattan distance between two observation

```python
a=df.iloc[5:6,:]
a
```

```python
a.values
```

```python
a.to_numpy()
```

```python
b=df.iloc[8:9,:]
b.to_numpy()
```

```python
abs(a-b).sum()
```

```python
a=df.iloc[5:6,:].values
b=df.iloc[8:9,:].values
abs(a-b).sum()
```

```python
a=df.iloc[5:6,:].to_numpy()
b=df.iloc[8:9,:].to_numpy()
abs(a-b).sum()
```

##  library for getting distance matrix for entire data(euclidean)

```python
from sklearn.metrics import pairwise_distances
X_precomputed = pairwise_distances(df,metric = 'euclidean')
df.head()
```

```python
df_2 = pd.DataFrame(X_precomputed)
df_2
```

## Manhatten matrix

```python
X_precomputed2 = pairwise_distances(df,metric= 'manhattan')
df_3 = pd.DataFrame(X_precomputed2)
df_3
```

# Handling Missing values

```python
# filling with mean
df.fillna(df['b'].mean(),inplace=True)
df
```

```python
df.fillna(df['b'].median(),inplace =True)
df
```

```python
df.fillna(df['a'].median(),inplace =True)
df
```

```python
df.fillna(method='ffill')
df.fillna(method ='bfill')
```
